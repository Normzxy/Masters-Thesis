```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_breast_cancer
```

```python
# data = load_digits()
data = load_breast_cancer()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target
# TO MARK MALIGNANT CASE AS 1 #
df['target'] = 1 - df['target']
```

```python
outlier_pcts = np.arange(1, 51, 1)
dfs = {0: df}

for pct in outlier_pcts:
    path = f"../../data/perturbed_datasets/BCW_1to50pct_15of30_g3.0/perturbed_{pct}_pct.csv"
    dfs[pct] = pd.read_csv(path)
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV, StratifiedKFold, PredefinedSplit
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.svm import SVC
```

```python
from utils import helper_functions as hf
import importlib
importlib.reload(hf)
```

```python
outer_cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=101)

metric_scores = []
for pct, df in dfs.items():

    X = df.drop('target', axis=1)
    Y = df['target']

    folds_metric_scores = []
    # Outer U and T split #
    for u_idx, t_idx in outer_cv.split(X, Y):
        X_u, Y_u = X.iloc[u_idx], Y.iloc[u_idx]
        X_t, Y_t = X.iloc[t_idx], Y.iloc[t_idx]

        # Inner X, V and T split #
        (X_x, X_v, Y_x, Y_v) = train_test_split(
            X_u, Y_u, test_size=0.2, stratify=Y_u, random_state=101)

        v_fold = np.concatenate([np.full(len(X_x), -1, dtype=int),
                                    np.zeros(len(X_v), dtype=int)])
        ps = PredefinedSplit(v_fold)

        # GridSearchCV expects (X, Y) as parameters #
        X_for_fit = pd.concat([X_x, X_v], axis=0)
        Y_for_fit = pd.concat([Y_x, Y_v], axis=0)

        # Scaling performed as a part of the process #
        pipe_steps = Pipeline([
        ('scaler', StandardScaler()),
        ('svm', SVC(random_state=101))])

        param_grid = [
            {'svm__kernel': ['linear'],
             'svm__C': np.logspace(-2, 2, 5),
             'svm__class_weight': [None, 'balanced']},

            {'svm__kernel': ['rbf'],
             'svm__C': np.logspace(-2, 2, 5),
             'svm__gamma': ['scale', 'auto', 0.01, 0.1, 1],
             'svm__class_weight': [None, 'balanced']}
        ]

        # Refit on whole outer training set U (X + V), to get the best model
        grid_model = GridSearchCV(pipe_steps, param_grid,
                            cv=ps, scoring='balanced_accuracy', n_jobs=-1, refit=True)
        grid_model.fit(X_for_fit, Y_for_fit)
        best_classifier = grid_model.best_estimator_

        this_fold_metrics = hf.evaluate_model(best_classifier, X_t, Y_t)
        folds_metric_scores.append(this_fold_metrics)

    fms = pd.DataFrame(folds_metric_scores)

    pct_summary = fms.mean(axis=0).to_dict()
    pct_summary['pct'] = pct
    metric_scores.append(pct_summary)
```

```python
ms = pd.DataFrame(metric_scores).set_index('pct')
ms.head(5)
```

```python
import matplotlib.pyplot as plt
plt.style.use('default')
```

```python
plt.figure(figsize=(10, 6), dpi=500)
# METRICS TO BE DRAWN #
metrics = ['f1_score', 'balanced_accuracy']
for metric in metrics:
    plt.plot(ms.index, ms[metric], label=metric)

metric_min = round(ms[metrics].min().min(), 2)
metric_max = round(ms[metrics].max().max(), 2)

plt.xticks(np.arange(ms.index.min(), ms.index.max() + 1, 1))
plt.xticks(rotation=60, size=8)
plt.yticks(np.arange(metric_min, metric_max + 0.01, 0.01))
plt.yticks(rotation=30, size=8)

plt.xlabel('Procent modyfikowanych wierszy', size=10)
plt.xlim(ms.index.min(), ms.index.max())
plt.ylabel('Wartość metryki', size=10)
# plt.title('Wydajność modelu w zależności od procentowego udziału wartości odstających')

plt.grid(True)
plt.legend(loc='upper right', prop={'size': 12}, bbox_to_anchor=(0.98, 0.96))
plt.tight_layout()
```
