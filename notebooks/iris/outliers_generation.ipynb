```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.neighbors import LocalOutlierFactor
from utils import generate_around_outliers
```

```python
####################
# DATA PREPARATION #
####################
```

```python
# DATA DIVISION AND CLARIFICATION #
data = load_iris()
df = pd.DataFrame(data.data, columns=data.feature_names)
df.rename(columns={'sepal length (cm)':'sepal_length',
                     'sepal width (cm)':'sepal_width',
                     'petal length (cm)':'petal_length',
                     'petal width (cm)':'petal_width'}, inplace=True)
df['target'] = data.target

X = df.drop('target', axis=1)
Y = df['target']
```

```python
# DATA SCALING #
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# MARKING THE 4% OF MOST OUTLYING POINTS (6 POINTS) #
lof_0 = LocalOutlierFactor(contamination=0.04, n_neighbors=10)
# OUTLIERS MARKED AS -1.
labels = lof_0.fit_predict(X_scaled)
# CREATE A COPY OF THE OUTLIERS AS A REFERENCE
outlier_loc = (df[labels == -1]
               .copy().reset_index(drop=True))
```

```python
# REFERENCE OUTLIERS #
outlier_loc
```

```python
################################################
# GENERATE NEW POINTS AROUND EXISTING OUTLIERS #
################################################
```

```python
# CUSTOM FUNCTION TO GENERATE NEW OUTLIERS BASED ON REFERENCE POINTS #
added_outliers = generate_around_outliers(
    original_data=X,
    input_outliers=outlier_loc,
    pct_to_enter=20,
    random_state=101,
    decimal_places=1)
```

```python
# CREATE A SAMPLE DATA FRAME WITH NEW OUTLIERS INCLUDED #
added_outliers_copy = added_outliers.copy()
df_copy = df.copy()

added_outliers_copy['is_added'] = 1
df_copy['is_added'] = -1

df_merged = pd.concat([df_copy, added_outliers_copy], ignore_index=True)
features = df_merged.drop(columns=['is_added', 'target'])
```

```python
# POST-CONCAT DATA SCALING #
df_merged_scaled = scaler.fit_transform(features)

# NOW 20% IS EXPECTED TO BE OUTLIERS #
lof_1 = LocalOutlierFactor(contamination=0.2, n_neighbors=10)
```

```python
# TO CHECK IF NEW OUTLIERS ARE DETECTED #
df_merged['outliers_detection'] = lof_1.fit_predict(df_merged_scaled)
df_merged[df_merged['is_added'] == 1]
```

```python
# HOW DO NEW OUTLIERS LOOK IN SPACE? #
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8), dpi=100)
sns.pairplot(df_merged.drop(['is_added', 'target'], axis=1),
             hue='outliers_detection', palette='managua')
```

```python
##########################################################
# GENERATE MANY FILES WITH NEW OUTLIERS, AND SAVE TO CSV #
##########################################################

percentages = np.arange(2, 52, 2)

for p in percentages:
    df_out = generate_around_outliers(
        original_data=X,
        input_outliers=outlier_loc,
        pct_to_enter=p, save=True,
        key_word="iris",
        random_state=101
    )
```
