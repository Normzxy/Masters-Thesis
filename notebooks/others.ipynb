```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
x = np.linspace(-3, 3, 1000)

pdf = norm.pdf(x, loc=0)

plt.figure(figsize=(10, 6), dpi=500)
plt.plot(x, pdf, label=r'$f(x) = \frac{1}{\sqrt{2\pi}} \, e^{-\frac{1}{2}x^2}$')
plt.title('Gęstość rozkładu normalnego w przedziale [-3, 3]')
plt.xlabel('x', size=16)
plt.xlim(-3, 3)
plt.ylabel('f(x)', size=16)
plt.grid(True)
plt.legend(prop={'size': 12}, bbox_to_anchor=(0.98, 0.85))
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
```

```python
# SOURCE: scikit-learn.org #
X, y = make_classification(random_state=0)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
clf = SVC(random_state=0)
clf.fit(X_train, y_train)
SVC(random_state=0)
predictions = clf.predict(X_test)
cm = confusion_matrix(y_test, predictions, labels=clf.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm,
                              display_labels=clf.classes_)
plt.figure(figsize=(10, 10), dpi=500)
disp.plot()
plt.show()
```

```python
def sigmoid(array):
    result = 1/(1 + np.exp(-array))
    return result

x = np.linspace(-6, 6)
y = sigmoid(x)

plt.figure(figsize=(10, 6), dpi=500)
plt.plot(x, y, label=r'$\sigma(x) =\; \frac{1}{1 + e^{-x}}$')
plt.xlabel('x', size=16)
plt.xlim(-6, 6)
plt.ylabel(r'$\sigma(x)$', size=16)
plt.ylim(-0.1, 1.1)
plt.grid(True)
plt.legend(loc='center right', prop={'size': 16}, bbox_to_anchor=(0.98, 0.5))
```
